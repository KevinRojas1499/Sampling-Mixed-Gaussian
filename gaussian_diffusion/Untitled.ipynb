```python
def dualPlot(samples1, samples2):
    plt.subplot(1, 2, 1) # row 1, col 2 index 1
    plt.title("Samples Before FFT")
    plt.scatter(samples1[:,0],samples1[:,1],color='red')


    plt.subplot(1, 2, 2) # index 2
    plt.title("Samples after FFT")
    plt.scatter(samples2[:,0],samples2[:,1],color='blue')

    plt.show()
    
def plot3D(samples,ax,title):
    ax.scatter3D(samples[:,0], samples[:,1], samples[:,2], color = "green")
    plt.title(title)
    
    
def dual3DPlot(samples1, samples2):
    fig = plt.figure()
    ax = fig.add_subplot(1, 2, 1, projection='3d')    
    plot3D(samples1, ax,"Samples Before FFT")

    ax = fig.add_subplot(1, 2, 2, projection='3d')    
    plot3D(samples2, ax, "Samples after FFT")
    
    plt.show()    
```

```python
num_samples = 1000

c = [1/3,1/4,1/5,1/6,1/20]
means = [[0.5,0.5],[-15,-20], [30,10], [7,15], [-6, 15]]
variances = [[[1,0],[0,1]], [[5,1],[1,1]] , [[1, 2],[2,5]], [[-1,-2],[-2,-5]],[[1,2],[2,4]]]


sde = sde_lib.LinearSDE(beta=20)
samplesBeforeFFT = torch.tensor(generateSamples.get_samples_from_mixed_gaussian(c,means,variances,num_samples))
samples = samplesBeforeFFT
samplesAfterFFT = torch.fft.fft(samplesBeforeFFT,norm="forward")

samplesFor2DFFT = torch.fft.fft2(samplesBeforeFFT,norm="forward")


dualPlot(samples,samplesAfterFFT)
dualPlot(samples,samplesFor2DFFT)
```
